1. Get Data Frame 

```python
import pandas as pd
import numpy as np
import requests
```

```python
PC_url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
PC_page = requests.get(PC_url)
PC_tables = pd.read_html(PC_page.text)
```

2. Create DF

```python
df_pc = PC_tables[0]
df_pc
```

3. Preprocessing

```python
df_pc = df_pc.loc[1:180]
df_pc = df_pc.dropna(how='any')
```

```python
df_pc.columns = ['Postal code','Borough','Neighborhood']
```

```python
df_pc.head()
```

```python
grouped = df_pc['Neighborhood'].groupby(df_pc['Postal code'])
grouped.sum()
```

```python
range1 = range(0, 103)
```

```python
range1 = pd.Series(range1)
```

```python
df_pc = df_pc.set_index(range1)
df_pc.head()
```
